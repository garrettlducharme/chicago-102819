```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = np.random.normal(size = 100)
```

```python
plt.hist(data, bins = 20)
plt.title('Histogram of 100 standard normal samples')
plt.savefig('visualizations/normal_hist.png')
```

```python
sns.distplot(data, bins = 20)
plt.title('Histogram of 100 standard normal sample with kde')
plt.savefig('visualizations/normal_hist_kde.png')
```

```python
x = np.random.normal(0,1,100)*0.3
y = x + np.random.normal(0,0.1,100)

sns.regplot(x,y);
```
