```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
normal = np.random.normal(0,1, size =100)
```

```python
plt.hist(normal)
```

```python
sns.distplot(normal)
```

```python
noise = np.random.normal(0,.1,100)
```

```python
weight = .5
```

```python
y = normal*weight + noise
```

```python
sns.regplot(normal, y)
```
